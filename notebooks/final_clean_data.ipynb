```python
import pandas as pd

df = pd.read_csv('../data/cleaned_austin_v2.csv', low_memory=False)

# Show counts for each property type BEFORE one-hot encoding
property_cols = [c for c in df.columns if c.startswith('property_type_')]
property_counts = df[property_cols].sum().sort_values(ascending=False)

print(property_counts)
```

```python
property_cols = [c for c in df.columns if c.startswith('property_type_')]

entire_cols = [c for c in property_cols if 'entire' in c.lower()]
private_cols = [c for c in property_cols if 'private room' in c.lower() or 'shared room' in c.lower()]
hotel_cols = [c for c in property_cols if 'hotel' in c.lower()]

df['is_entire_place'] = df[entire_cols].sum(axis=1).clip(0,1)
df['is_private_room'] = df[private_cols].sum(axis=1).clip(0,1)
df['is_hotel_room'] = df[hotel_cols].sum(axis=1).clip(0,1)

df = df.drop(columns=property_cols)
```
