```python
import pandas as pd
```

```python
df = pd.read_csv('../data/cleaned_austin_v1.csv', low_memory=False)

print("Original Shape:", df.shape)
```

```python
# Drop misleading / low-value / noisy features
drop_cols = [
    'has_free_parking',
    'missing_response_rate',
    'missing_acceptance_rate',
    'host_response_rate',
    'review_scores_cleanliness',
    'review_scores_communication',
    'review_scores_location'
]

df = df.drop(columns=[c for c in drop_cols if c in df.columns])
print("After dropping noisy columns", df.shape)
```

```python
# Clip unrealistic values
df['minimum_nights'] = df['minimum_nights'].clip(1, 30)
df['maximum_nights'] = df['maximum_nights'].clip(1, 365)

df['reviews_per_month'] = df['reviews_per_month'].clip(0, 6)
df['price'] = df['price'].clip(20, df['price'].quantile(0.99))
```

```python
df.to_csv("../data/cleaned_austin_v2.csv", index=False)
```
